# DATA VISUALIZATION



# 1.The best selling books of the decade.

To get the best selling book of the year,I used the user rating to predict that

```python
data[['Name','User Rating']].nlargest(12,'User Rating').plot.bar(x='Name',y='User Rating',color='red')
plt.title('The best selling books of the decade.',size=15)
plt.xlabel('Name of the Books',fontsize=13,color='green')
plt.ylabel('Ratings',size=13,color='green')
plt.show()
```

# 2.The percentage of fiction to non-fiction.

```python
data['Genre'].value_counts()
```

```python
#Pie charts are generally used to show percentage or proportional data and...
#usually the percentage represented by each category is provided next to the corresponding slice of pie
```

```python
slices = [240, 310]
labels = [ 'Fiction','Non Fiction',]
colors = ['b', 'g']
plt.pie(slices, labels=labels, colors=colors, startangle=90, autopct='%.1f%%')
plt.title('Percentage of Fiction to Non Fiction')
plt.show()
```

# 3.Trends over the years.

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Trends Over The Years.")

sns.lineplot(data=data)
```

# 4.Top 10 Authors according to reviews.

```python
data[['Author','Reviews']].nlargest(10,'Reviews').plot.bar(x='Author',y='Reviews',color='blue')
plt.title('Top 10 Authors according to reviews',size=15)
plt.xlabel('Authors',size=15,color='red')
plt.ylabel('Number of reviews',size=15, color='green')
plt.show()
```
